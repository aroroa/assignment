```python
import csv
from heap import Heap
from datetime import datetime

heap = Heap()

with open('birthday.csv', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        if len(row) < 3 or row[2].strip() == '':
            continue
        name = row[1]
        birthday_str = row[2].strip()
        try:
            birthday = datetime.strptime(birthday_str, "%Y%m%d")
            heap.insert((birthday, name))
        except ValueError:
            continue

print("생일이 느린 순서로 10명의 친구:")
for _ in range(10):
    if not heap.isEmpty():
        birthday, name = heap.deleteMax()
        print(f"{name}: {birthday.strftime('%Y-%m-%d')}")
```

#4. 같은 조 친구들의 이름과 생년월일 출력

```python
import csv
from circularDoublyLinkedList import CircularDoublyLinkedList

my_team_names = ['김연진','변수연','정윤서','박서연','노은서','오세은','박성연','김민경','김보민','홍서연']

cdll = CircularDoublyLinkedList()

with open('birthday.csv', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        hakbun = row[0].strip()
        name = row[1].strip()
        birthday = row[2].strip()
        cdll.append((hakbun, name, birthday))

print("같은 조 친구들:")
for item in cdll:
    hakbun, name, birthday = item
    if name in my_team_names and (name != '홍서연' or hakbun.endswith('14')):
        print(f"{name} : {birthday[:4]}-{birthday[4:6]}-{birthday[6:]}")

```

# 6.LeetCode 703.Kth Largest Element in Stream

```python
import heapq

class KthLargest:

    def __init__(self, k, nums):
        self.k = k
        self.min_heap = nums
        heapq.heapify(self.min_heap)
        while len(self.min_heap) > k:
            heapq.heappop(self.min_heap)

    def add(self, val):
        if len(self.min_heap) < self.k:
            heapq.heappush(self.min_heap, val)
        elif val > self.min_heap[0]:
            heapq.heappushpop(self.min_heap, val)
        return self.min_heap[0]
```
